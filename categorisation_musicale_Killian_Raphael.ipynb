#### On vérifie la corrélation entre tout les paramètres, afin de déterminer lesquels sont les plus importants pour que le modèle soit le plus précie possible et ainsi réduire la dimension des données à interpréter.

```python
from scipy import stats
columns = ['chroma_0_mean', 'chroma_0_std', 'chroma_10_mean', 'chroma_10_std', 'chroma_11_mean', 
           'chroma_11_std', 'chroma_1_mean', 'chroma_1_std', 'chroma_2_mean', 'chroma_2_std', 'chroma_3_mean', 
           'chroma_3_std', 'chroma_4_mean', 'chroma_4_std', 'chroma_5_mean', 'chroma_5_std', 'chroma_6_mean', 
           'chroma_6_std', 'chroma_7_mean', 'chroma_7_std', 'chroma_8_mean', 'chroma_8_std', 'chroma_9_mean', 
           'chroma_9_std', 'chroma_centroid_mean', 'chroma_flux_mean', 'chroma_max', 'chroma_min', 'chroma_spread_mean', 
           'mfcc_10_mean', 'mfcc_10_std', 'mfcc_11_mean', 'mfcc_11_std', 'mfcc_12_mean', 'mfcc_12_std', 'mfcc_13_mean', 
           'mfcc_13_std', 'mfcc_2_mean', 'mfcc_2_std', 'mfcc_3_mean', 'mfcc_3_std', 'mfcc_4_mean', 'mfcc_4_std', 'mfcc_5_mean', 
           'mfcc_5_std', 'mfcc_6_mean', 'mfcc_6_std', 'mfcc_7_mean', 'mfcc_7_std', 'mfcc_8_mean', 'mfcc_8_std', 'mfcc_9_mean', 
           'mfcc_9_std', 'onset_rate', 'spectral_centroid_mean', 'spectral_decrease_mean', 'spectral_flux_mean', 
           'spectral_rolloff_mean', 'spectral_spread_mean', 'zcr_mean', 'zcr_std']
Y = Label
res = []
for j in range(len(columns)-1):
    slope, intercept, r_value, p_value, _ = stats.linregress(X_train[columns[j]],Y)
    
    print('modele ', j+1)
    print('----- Valeur Totale = ', intercept, ' + ', slope,' * ', columns[j])
    print('----- R² = ', r_value)
    print('      Donc {:2.2f}% de la variabilité de la valeur totale est expliquée par {:s}'.format(r_value *100,columns[j]))
    print('----- p value = ', p_value)
    print('      H0: beta1 = 0')
    
    if p_value < 0.05 :
        print('      ',p_value,'<=5%','==> on rejette H0, au seuil de risque alpha = 5%')
        print('          le modèle ',j+1, 'est bien pertinent.')
        res.append(columns[j])
    else : 
        print('      ',p_value,'> 5%','==> on accepte H0, au seuil de risque alpha = 5%')
        print('          le modèle ',j+1, 'n est pas pertinent car pas statistiquement significatif.')
print(res)
X = X_train[res]

len(res)
```

#### Nous avons 33 paramètres déterminants

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=33)
pca.fit(X_train)
```

```python
X_train_reduc = pca.transform(X_train)
print(X_train_reduc)
```

```python
print(pca.explained_variance_ratio_)
```

```python
result = 0
for ratio in pca.explained_variance_ratio_:
    result+=ratio
print(result)
```

#### Nous avons transformer les paramètres en pourcentage afin de calculer le ratio de précision, notre PCA est donc de 97,9% <br><br> Ensuite nous utilisons train_test_split afin de créer un échantillon non biaisé.

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, Y, random_state = 0)
print(X_train, X_val, y_train, y_val)
```

### Modele SVM
#### Il s'agit de notre algorithme de classification

```python
from sklearn.svm import SVC
from sklearn import metrics

svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)

y_pred = svclassifier.predict(X_val)
```

### Accuracy & Matrice de confusion
#### Performance de notre modele = 80%

```python
scores = metrics.accuracy_score(y_val, y_pred)
print('Accuracy: ','{:2.2%}'.format(scores))

cm = metrics.confusion_matrix(y_val, y_pred)
print(cm)
```

#### On a comparé notre algorithme SVM avec un KNN, dont l'accuracy est respectivement 80% contre 72,5% <br> Le modele SVM est donc plus précis

```python
from sklearn.neighbors import KNeighborsClassifier
knn2 = KNeighborsClassifier()
knn2.fit(X_train, y_train)

y_pred2 = knn2.predict(X_val)

from sklearn.metrics import accuracy_score
acc = accuracy_score(y_val, y_pred2)
print('Accuracy: {:2.2%}'.format(acc))

def best_model_search(X_train,X_val,y_train,y_val, n=10, patience=3):
    k=1
    best_acc = 0
    best_model=None
    best_k=1
    while k<n and patience>0:
        knn =KNeighborsClassifier(n_neighbors = k)
        knn.fit(X_train,y_train)
        y_pred = knn.predict(X_val)
        accuracy = accuracy_score(y_val, y_pred)
        print('accuracy : {:.2%} '.format(accuracy),'avec k = ',k)
        if accuracy > best_acc:
            best_acc = accuracy
            best_model = knn
            best_k=k
        patience = patience-1
        k=k+1
    return best_model,best_k,best_acc

knn2, k, acc = best_model_search(X_train,X_val,y_train,y_val,n=100, patience=30)
print('best_model = ',knn2,' accuracy : {:.2%} '.format(acc),'Pour un k = ',k)
```
